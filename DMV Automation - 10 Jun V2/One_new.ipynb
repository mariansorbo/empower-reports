```python
import subprocess
import sys
import os
import config_runtime
import config
import pyautogui
import time
from pywinauto import Application, Desktop

# â€”â€”â€”â€”â€”â€” ConfiguraciÃ³n de la salida safe_print â€”â€”â€”â€”â€”â€”
if hasattr(sys.stdout, "reconfigure"):
    sys.stdout.reconfigure(encoding="utf-8", errors="replace")

def safe_print(text):
    try:
        print(text)
    except UnicodeEncodeError:
        enc = sys.stdout.encoding or "utf-8"
        fixed = text.encode(enc, errors="replace").decode(enc)
        print(fixed)

# ðŸ“‚ Cargar configuraciÃ³n
usuario = config_runtime.usuario
# reporte puede venir sin extensiÃ³n
reporte = config_runtime.reporte.strip()
if not reporte.lower().endswith(".pbit"):
    reporte += ".pbit"

base_report_directory = config.base_report_directory

# â€”â€”â€”â€”â€”â€” Buscar el archivo real en la carpeta â€”â€”â€”â€”â€”â€”
base_name = os.path.splitext(reporte)[0]
candidatos = {
    f"{base_name} + {usuario}.pbit",
    f"{usuario} + {base_name}.pbit"
}

found_file = None
for f in os.listdir(base_report_directory):
    if f in candidatos:
        found_file = f
        break

if not found_file:
    safe_print(f"âŒ No encontrÃ© ningÃºn .pbit en {base_report_directory} que sea:\n  â€¢ {base_name} + {usuario}.pbit\n  â€¢ {usuario} + {base_name}.pbit")
    safe_print("Archivos .pbit disponibles:")
    for f in os.listdir(base_report_directory):
        if f.lower().endswith(".pbit"):
            safe_print(f"  â€“ {f}")
    sys.exit(1)

pbit_file_path = os.path.join(base_report_directory, found_file)
safe_print(f"ðŸ“Š Analizando el reporte real: {found_file}  | Usuario: {usuario}")

# â€”â€”â€”â€”â€”â€” Construir y lanzar Power BI â€”â€”â€”â€”â€”â€”
powerbi_exe_path = config.powerbi_exe_path
command_line = f'"{powerbi_exe_path}" "{pbit_file_path}"'

safe_print("ðŸŸ¡ Lanzando Power BI...")
app = Application(backend="uia").start(command_line)
safe_print("ðŸŸ¢ Power BI lanzado.")
time.sleep(10)

# â€”â€”â€”â€”â€”â€” Maximizar ventana de Power BI â€”â€”â€”â€”â€”â€”
try:
    safe_print("ðŸ” Buscando ventana principal de Power BI para maximizar...")
    max_window = None
    for _ in range(10):
        for w in Desktop(backend="uia").windows():
            if "Power BI" in w.window_text():
                max_window = w
                break
        if max_window:
            break
        time.sleep(1)

    if max_window:
        max_window.set_focus()
        max_window.maximize()
        safe_print("ðŸŸ¢ Power BI maximizado correctamente.")
    else:
        safe_print("âš ï¸ No se encontrÃ³ la ventana principal de Power BI para maximizar.")
except Exception as e:
    safe_print(f"âŒ Error al maximizar Power BI: {e}")

time.sleep(1.5)
pyautogui.click(725, 1201)
```

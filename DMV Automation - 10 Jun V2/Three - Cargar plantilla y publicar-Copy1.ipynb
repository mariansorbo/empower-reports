# ðŸ“˜ DocumentaciÃ³n TÃ©cnica â€“ PublicaciÃ³n Automatizada de Reporte en Power BI
ðŸ§  Objetivo del Script
Este script automatiza completamente el proceso de publicaciÃ³n de un archivo .pbix en Power BI Service, partiendo desde su apertura local con Power BI Desktop, refrescando los datos y finalizando con la publicaciÃ³n y la copia del link pÃºblico del reporte publicado.

ðŸ“ Estructura y Funcionamiento
1. InicializaciÃ³n y Apertura
Se utiliza pyautogui para emular clicks y escritura.

El script inicia haciendo clic en la barra de bÃºsqueda de Windows y escribiendo "power" para lanzar Power BI Desktop.

Luego, selecciona y abre la Plantilla Empower BI, que se espera estÃ© visible en la lista de recientes de Power BI.

2. PublicaciÃ³n del Reporte
Una vez abierta la plantilla, el script:

Hace clic en el botÃ³n "Publicar".

InteractÃºa con el menÃº desplegable del workspace.

Selecciona el workspace deseado.

Confirma la publicaciÃ³n.

ðŸ“ Nota tÃ©cnica: El proceso se basa en coordenadas pregrabadas. Requiere una resoluciÃ³n de pantalla estable y sin interrupciones.

-------------------------------------------------

ðŸŒ RecuperaciÃ³n del Link de Reporte
El script continÃºa:

Abriendo el reporte publicado (desde el menÃº emergente).

Haciendo clic en la barra de direcciÃ³n del navegador.

Copiando la URL pÃºblica usando el portapapeles.

Imprimiendo ese link como salida para el usuario.

-------------------------------------------------

ðŸ—‚ï¸ Origen de los Datos
Los datos que alimentan la plantilla:

Fueron previamente refrescados y guardados en OneDrive y Google Drive por un script anterior.

La plantilla de Power BI estÃ¡ configurada para apuntar a ese archivo .xlsx vÃ­a OneDrive, permitiendo su actualizaciÃ³n automÃ¡tica al abrir.

-------------------------------------------------


ðŸ› ï¸ Dependencias y Herramientas
pyautogui: Para automatizar clicks, escritura y movimientos.

time: Para controlar los delays entre pasos (espera de carga).

pynput.mouse: Para capturar coordenadas de click en sesiones de entrenamiento manual del script.

-------------------------------------------------


ðŸ§© Requisitos Previos
El archivo .pbix debe estar ubicado en:
C:\Users\Administrator\OneDrive\Empower BI Archivos\Plantilla Empower BI.pbix

El archivo Excel con los datos debe haberse generado y subido correctamente antes de ejecutar este script.

Power BI Desktop debe estar instalado, accesible y su interfaz debe mantenerse estable durante el proceso.

```python
import pyautogui
import pyperclip
import time
from pywinauto import Application, Desktop
import config
import config_runtime

usuario = config_runtime.usuario
carpeta_destino = config.step4_entregables

# Conectarse a Power BI
app = Application(backend="uia").connect(path="PBIDesktop.exe")
main_window = app.top_window()
main_window.set_focus()

# Esperar que aparezca ventana 'Save As'
print("[ðŸ“] Esperando ventana 'Save As' (como hijo)...")
save_as_window = None

# Reintentar por hasta 5 segundos
for _ in range(10):
    for child in main_window.children():
        if (
            child.element_info.control_type == "Window"
            and child.window_text().strip().lower() == "save as"
            and child.element_info.class_name == "#32770"
        ):
            save_as_window = app.window(handle=child.handle)  # ðŸ‘ˆ Esta lÃ­nea lo soluciona
            break
    if save_as_window:
        break
    time.sleep(0.5)

if not save_as_window:
    print("âŒ No se abriÃ³ la ventana 'Save As'.")
else:
    try:
        save_as_window.set_focus()
        print("[âœ“] Ventana 'Save As' detectada.")

        # ðŸ”¤ 1. Escribir nombre del archivo
        nombre_archivo = f"{usuario} - Empower BI"
        pyperclip.copy(nombre_archivo)
        name_input = save_as_window.child_window(auto_id="1001", control_type="Edit")
        name_input.set_focus()
        name_input.type_keys("^a{BACKSPACE}", with_spaces=True)
        pyautogui.hotkey("ctrl", "v")
        print(f"[âœ“] Nombre de archivo seteado: {nombre_archivo}")

        # ðŸ“‚ 2. Cambiar carpeta destino
        pyperclip.copy(carpeta_destino)
        pyautogui.hotkey("ctrl", "l")  # activar path input
        time.sleep(0.3)
        pyautogui.hotkey("ctrl", "v")
        pyautogui.press("enter")
        time.sleep(1.2)
        print(f"[âœ“] Carpeta destino seteada: {carpeta_destino}")

        # ðŸ” 3. Refocus a ventana 'Save As' por si redibuja
        # ðŸ” 3. Refocus a ventana 'Save As' por si redibuja
        for child in main_window.children():
            if (
                child.window_text().strip().lower() == "save as"
                and child.element_info.control_type == "Window"
                and child.element_info.class_name == "#32770"
            ):
                save_as_window = app.window(handle=child.handle)
                save_as_window.set_focus()
                break

        # ðŸ’¾ 4. Click en botÃ³n 'Save'
        save_button = save_as_window.child_window(title="Save", control_type="Button")
        save_button.click_input()
        print("[âœ“] Click en 'Save' exitoso.")
        time.sleep(2)

        # ðŸ” 5. Confirmar reemplazo si aparece
        for c in main_window.descendants():
            if c.window_text().strip().lower() == "yes" and c.element_info.control_type == "Button":
                c.click_input()
                print("[â†ª] ConfirmaciÃ³n de reemplazo aceptada.")
                break

        # âœ… VerificaciÃ³n final
        time.sleep(1)
        still_open = any(
            child.window_text().strip().lower() == "save as" and child.element_info.control_type == "Window"
            for child in main_window.children()
        )
        if not still_open:
            print(f"[âœ…] Archivo '{nombre_archivo}' guardado correctamente.")
        else:
            print(f"[âŒ] FallÃ³ el guardado de '{nombre_archivo}'.")

    except Exception as e:
        print(f"âŒ Error al manejar 'Save As': {e}")
time.sleep(0.3)
```

```python
from pywinauto import Application, Desktop
import time

# Conectarse a Power BI
app = Application(backend="uia").connect(path="PBIDesktop.exe")
main_window = app.top_window()
main_window.set_focus()

print("ðŸ” Buscando ventanas hijas del proceso Power BI...")
time.sleep(2)  # esperar a que se renderice

# Buscar Dialog
dialog = None
for child in main_window.children():
    if child.friendly_class_name() == "Dialog":
        dialog = child
        break

if not dialog:
    print("âŒ No se encontrÃ³ ninguna ventana tipo Dialog.")
else:
    print("âœ… Ventana Dialog detectada.")

    try:
        # Buscar y clickear el ListItem 'dgh'
        list_item = next(
            (c for c in dialog.descendants()
             if c.friendly_class_name() == "ListItem" and c.window_text().strip().lower() == "dgh"),
            None
        )
        if list_item:
            list_item.click_input()
            print("âœ… Workspace 'dgh' seleccionado.")
        else:
            print("âŒ No se encontrÃ³ el ListItem 'dgh'.")

        time.sleep(0.5)

        # Buscar y clickear el botÃ³n 'Select'
        select_btn = next(
            (c for c in dialog.descendants()
             if c.friendly_class_name() == "Button" and c.window_text().strip().lower() == "select"),
            None
        )
        if select_btn:
            select_btn.click_input()
            print("âœ… BotÃ³n 'Select' clickeado.")
        else:
            print("âŒ No se encontrÃ³ el botÃ³n 'Select'.")

        # Esperar posible pop-up de reemplazo
        print("â³ Esperando posible ventana de reemplazo...")
        time.sleep(2)
        for i in range(10):
            for d in main_window.children():
                if d.friendly_class_name() == "Dialog":
                    replace_btn = next(
                        (c for c in d.descendants()
                         if c.friendly_class_name() == "Button" and c.window_text().strip().lower() == "replace"),
                        None
                    )
                    if replace_btn:
                        replace_btn.click_input()
                        print("âœ… BotÃ³n 'Replace' clickeado.")
                        break
            else:
                time.sleep(0.5)
                continue
            break

        # Esperar mensaje de Ã©xito y click en el link
        print("â³ Esperando mensaje 'Success!' y link para abrir en Power BI...")
        for _ in range(30):
            for d in main_window.children():
                if d.friendly_class_name() == "Dialog":
                    success_link = next(
                        (c for c in d.descendants()
                         if c.window_text().strip().lower().startswith("open")
                         and "in power bi" in c.window_text().lower()),
                        None
                    )
                    if success_link:
                        success_link.click_input()
                        print(f"âœ… Link '{success_link.window_text()}' clickeado.")
                        break
            else:
                time.sleep(0.5)
                continue
            break
        else:
            print("âš ï¸ No se detectÃ³ el link de apertura tras la publicaciÃ³n.")

    except Exception as e:
        print(f"âŒ Error durante la automatizaciÃ³n: {e}")
```

```python
import pandas as pd
import time
import pyperclip
from pywinauto import Application
from pywinauto.mouse import click
from pywinauto.keyboard import send_keys

# Ruta del archivo
file_path = r"C:\Users\Administrator\DMV Automation - 10 Jun\Archivos Consolidados\testUser - Consolidado.xlsx"

# Leer hoja
df = pd.read_excel(file_path, sheet_name="Columnas y Métricas")
df["Formatted Expression"] = ""  # Nueva columna

# Coordenadas (las que pasaste)
EDITOR_COORD = (1581, 1387)
FORMAT_BUTTON_COORD = (1574, 1205)
TEXT_RESULT_COORD = (1581, 1387)  # donde hacemos clic para seleccionar texto formateado

# Conectar a DAX Studio
app = Application(backend="uia").connect(title_re=".*DAX Studio.*")
win = app.window(title_re=".*DAX Studio.*")
win.set_focus()

# Iterar expresiones
for i, expr in enumerate(df["Expression"].dropna(), start=0):
    print(f"[{i+1}] Formateando expresión...")

    # Copiar al portapapeles
    pyperclip.copy(str(expr))
    time.sleep(0.3)

    # Pegar en editor
    click(coords=EDITOR_COORD)
    time.sleep(0.1)
    send_keys("^a{BACKSPACE}")  # Limpiar
    time.sleep(0.1)
    send_keys("^v")  # Pegar
    time.sleep(0.1)

    # Clic en 'Format Query'
    click(coords=FORMAT_BUTTON_COORD)
    time.sleep(0.35)

    # Clic donde aparece el texto formateado
    click(coords=TEXT_RESULT_COORD)
    time.sleep(0.1)
    send_keys("^a")  # Seleccionar todo
    time.sleep(0.1)
    send_keys("^c")  # Copiar
    time.sleep(0.1)

    # Leer resultado
    formatted = pyperclip.paste()
    df.at[i, "Formatted Expression"] = formatted
    print(f"[{i+1}] ✅ OK")

# Guardar
output_path = file_path.replace(".xlsx", "_withFormattedLong.xlsx")
df.to_excel(output_path, index=False)
print(f"\n✅ Guardado como: {output_path}")
```

```python
import pandas as pd
import time
import pyperclip
import pyautogui
import keyboard
from pywinauto import Application
from pywinauto.mouse import click
from pywinauto.keyboard import send_keys

# --- 1. Configuración ---
file_path = r"C:\Users\Administrator\DMV Automation - 10 Jun\Archivos Consolidados\testUser - Consolidado.xlsx"
sheet_name = "Columnas y Métricas"

EDITOR_COORD = (1581, 1387)
TEXT_RESULT_COORD = (1581, 1387)

# --- 2. Leer Excel ---
df = pd.read_excel(file_path, sheet_name=sheet_name)
df["Formatted Expression"] = ""

# --- 3. Conectar a DAX Studio ---
app = Application(backend="uia").connect(title_re=".*DAX Studio.*")
win = app.window(title_re=".*DAX Studio.*")
win.set_focus()
time.sleep(0.5)

# --- 4. Loop principal ---
for i, expr in enumerate(df["Expression"].dropna(), start=0):
    if keyboard.is_pressed('p'):
        print("\n⛔️ Proceso abortado manualmente con tecla 'P'")
        break

    print(f"[{i+1}] Formateando expresión...")

    # Copiar al portapapeles
    pyperclip.copy(str(expr))
    time.sleep(0.05)

    # Click en editor y pegar
    click(coords=EDITOR_COORD)
    send_keys("^a{BACKSPACE}")
    time.sleep(0.05)
    send_keys("^v")

    # Ctrl+F6 para formatear
    pyautogui.hotkey('ctrl', 'f6')
    time.sleep(0.5)

    # Copiar texto formateado
    click(coords=TEXT_RESULT_COORD)
    pyautogui.hotkey('ctrl', 'a')
    pyautogui.hotkey('ctrl', 'c')

    # Leer texto formateado
    formatted = pyperclip.paste()
    df.at[i, "Formatted Expression"] = formatted
    print(f"[{i+1}] ✅ OK")

# --- 5. Guardar si se completó al menos una fila ---
if "Formatted Expression" in df.columns and df["Formatted Expression"].str.len().gt(0).any():
    output_path = file_path.replace(".xlsx", "_withFormattedShort.xlsx")
    df.to_excel(output_path, index=False)
    print(f"\n✅ Archivo guardado: {output_path}")
else:
    print("⚠️ No se guardó ningún cambio.")
```

```python
import pandas as pd
import pyperclip
import pyautogui
import keyboard
from pywinauto import Application
from pywinauto.mouse import click
from pywinauto.keyboard import send_keys
from time import sleep

# --- Configuración ---
file_path = r"C:\Users\Administrator\DMV Automation - 10 Jun\Archivos Consolidados\testUser - Consolidado.xlsx"
sheet_name = "Columnas y Métricas"
EDITOR_COORD = (1581, 1387)
TEXT_RESULT_COORD = (1581, 1387)

# --- Leer archivo ---
df = pd.read_excel(file_path, sheet_name=sheet_name)
df["Expression Formatted Short"] = ""
df["Expression Formatted Long"] = ""

# --- Conectar a DAX Studio ---
app = Application(backend="uia").connect(title_re=".*DAX Studio.*")
win = app.window(title_re=".*DAX Studio.*")
win.set_focus()
sleep(0.2)

# --- Loop ---
for i, expr in enumerate(df["Expression"].dropna(), start=0):
    if keyboard.is_pressed('p'):
        print("\n⛔️ Aborto manual con 'P'")
        break

    print(f"[{i+1}] Formateando...")

    pyperclip.copy(str(expr))
    sleep(0.05)

    click(coords=EDITOR_COORD)
    send_keys("^a{BACKSPACE}")
    sleep(0.05)
    send_keys("^v")
    sleep(0.05)

    # --- SHORT (Ctrl + F6) ---
    pyautogui.hotkey('ctrl', 'f6')
    sleep(0.4)
    click(coords=TEXT_RESULT_COORD)
    pyautogui.hotkey('ctrl', 'a')
    pyautogui.hotkey('ctrl', 'c')
    sleep(0.05)
    df.at[i, "Expression Formatted Short"] = pyperclip.paste()

    # --- LONG (F6) ---
    pyautogui.press('f6')
    sleep(0.4)
    click(coords=TEXT_RESULT_COORD)
    pyautogui.hotkey('ctrl', 'a')
    pyautogui.hotkey('ctrl', 'c')
    sleep(0.05)
    df.at[i, "Expression Formatted Long"] = pyperclip.paste()

    print(f"[{i+1}] ✅ OK")

# --- Guardar en el mismo archivo ---
with pd.ExcelWriter(file_path, engine="openpyxl", mode="a", if_sheet_exists="overlay") as writer:
    df.to_excel(writer, sheet_name=sheet_name, index=False)

print(f"\n✅ Guardado en archivo original: {file_path}")
```

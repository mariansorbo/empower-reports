```python
from pywinauto import Application, Desktop
import time
import os
import config_runtime
import config

usuario = config_runtime.usuario
reporte = config_runtime.reporte.strip()
powerbi_exe_path = config.powerbi_exe_path
base_report_directory = config.base_report_directory
carpeta_destino = config.step1_extracciones_dmv

# Asegurar extensi√≥n .pbit
if not reporte.endswith(".pbit"):
    reporte += ".pbit"

print("Analizando el reporte: " + reporte + " | Usuario: " + usuario)

pbit_file_path = os.path.join(base_report_directory, reporte)
command_line = f'"{powerbi_exe_path}" "{pbit_file_path}"'

# Lanzar Power BI
app = Application(backend="uia").start(command_line)
print("[‚úì] Power BI lanzado")
time.sleep(10)

# üî¥ Cerrar hasta 10 veces el bot√≥n 'Cancel' con clase formulaBarCancel...
try:
    powerbi_window = None
    for w in Desktop(backend="uia").windows():
        if "Power BI" in w.window_text():
            powerbi_window = w
            break

    if not powerbi_window:
        raise Exception("‚ùå No se encontr√≥ la ventana principal de Power BI.")

    close_count = 0
    for _ in range(10):
        found = False
        for ctrl in powerbi_window.descendants():
            if (ctrl.element_info.control_type == "Button"
                and ctrl.window_text().strip().lower() == "cancel"
                and ctrl.element_info.class_name.startswith("formulaBarCancel")):
                try:
                    ctrl.click_input()
                    close_count += 1
                    print(f"[‚úñ] Bot√≥n 'Cancel' (formulaBarCancel...) clickeado ({close_count}).")
                    found = True
                    time.sleep(0.5)
                except Exception as e:
                    print(f"‚ö†Ô∏è Fall√≥ intento de clic en 'Cancel': {e}")
        if not found:
            break
        time.sleep(1)

    if close_count == 0:
        print("‚ÑπÔ∏è No se encontr√≥ ning√∫n bot√≥n 'Cancel' con clase 'formulaBarCancel...'.")
except Exception as e:
    print(f"‚ùå Error al cerrar el bot√≥n 'Cancel': {e}")

# üß© Manejar ventana de par√°metros de plantilla
try:
    dlg = None
    for w in Desktop(backend="uia").windows():
        if "Enter Parameters" in w.window_text() or "Import" in w.window_text():
            dlg = w
            break

    if dlg:
        dlg.set_focus()
        print("[‚úì] Ventana de plantilla detectada:", dlg.window_text())

        btn = dlg.child_window(title_re="OK|Load|Aceptar|Cargar", control_type="Button")
        btn.click_input()
        print("[‚úì] Par√°metros confirmados y plantilla cargada.")
    else:
        print("‚ö†Ô∏è No apareci√≥ ventana de par√°metros. ¬øYa estaba precargada?")
except Exception as e:
    print("‚ùå Error al manejar la ventana de par√°metros:", e)
```

```python
import pyperclip
import pyautogui
import time

for query in queries:
    print(f"\nüîÅ Ejecutando query: {query['name']}")

    # 1. Click en editor de DAX
    pyautogui.click(1551, 1320)  # O us√° el control [161] si est√°s con pywinauto
    time.sleep(0.3)
    pyautogui.hotkey('ctrl', 'a')
    pyautogui.press('backspace')

    # 2. Pegar el query
    pyperclip.copy(query['sql'])
    pyautogui.hotkey('ctrl', 'v')
    time.sleep(0.3)

    # 3. Ejecutar query
    pyautogui.click(1137, 1207)  # Bot√≥n "Run"
    print("[‚úì] Query ejecutado.")
    time.sleep(1.5)

    # 4. Manejar ventana 'Save As'
    print("[üìÅ] Esperando ventana 'Save As'...")
    save_as_window = None

    for child in window.descendants():
        if child.element_info.control_type == "Window" and "Save As" in child.window_text():
            save_as_window = child
            break

    if not save_as_window:
        print("‚ùå No se abri√≥ la ventana 'Save As'.")
        continue

    try:
        save_as_window.set_focus()
        print(f"[‚úì] Ventana 'Save As' detectada.")

        # 4.1 Escribir nombre del archivo (usando pyperclip para evitar errores con '+')
        nombreArchivo = f"{usuario} + {nombre_reporte} + {query['name']}.xlsx"
        pyperclip.copy(nombreArchivo)
        name_input = save_as_window.descendants()[113]
        name_input.set_focus()
        name_input.type_keys("^a{BACKSPACE}", with_spaces=True)
        pyautogui.hotkey("ctrl", "v")
        print(f"[‚úì] Nombre de archivo: {nombreArchivo}")

        # 4.2 Cambiar carpeta destino
        pyperclip.copy(carpeta_destino)
        pyautogui.hotkey("ctrl", "l")
        time.sleep(0.3)
        pyautogui.hotkey("ctrl", "v")
        pyautogui.press("enter")
        time.sleep(1.2)
        print(f"[‚úì] Carpeta destino: {carpeta_destino}")

        # 4.3 Revalidar ventana 'Save As'
        for child in window.descendants():
            if child.element_info.control_type == "Window" and "Save As" in child.window_text():
                save_as_window = child
                save_as_window.set_focus()
                break

        # 4.4 Click en bot√≥n Save
        save_button = save_as_window.descendants()[123]
        save_button.click_input()
        print("[‚úì] Click en 'Save' exitoso.")
        time.sleep(2)

        # 4.5 Confirmar reemplazo si aparece
        updated_children = window.descendants()
        yes_button = next(
            (c for c in updated_children if c.element_info.control_type == "Button" and c.window_text().strip().lower() == "yes"),
            None
        )
        if yes_button:
            yes_button.click_input()
            print("[‚Ü™] Confirmaci√≥n de reemplazo aceptada.")

        # 5. Verificar si se cerr√≥ correctamente
        time.sleep(1)
        final_check = any(
            "Save As" in c.window_text() and c.element_info.control_type == "Window"
            for c in window.descendants()
        )
        if not final_check:
            print(f"[‚úÖ] Archivo '{nombreArchivo}' guardado correctamente.")
        else:
            print(f"[‚ùå] Fall√≥ el guardado de '{nombreArchivo}'.")

    except Exception as e:
        print(f"‚ùå Error durante manejo de 'Save As': {e}")
```

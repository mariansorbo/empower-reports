```python
# In[7]:


import subprocess
import sys
import os
import config_runtime
import config
import pyautogui
import time
from pywinauto import Application, Desktop

# ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî Configuraci√≥n de la salida safe_print ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
if hasattr(sys.stdout, "reconfigure"):
    sys.stdout.reconfigure(encoding="utf-8", errors="replace")

def safe_print(text):
    try:
        print(text)
    except UnicodeEncodeError:
        enc = sys.stdout.encoding or "utf-8"
        fixed = text.encode(enc, errors="replace").decode(enc)
        print(fixed)

# üìÇ Cargar configuraci√≥n
usuario = config_runtime.usuario
# reporte puede venir sin extensi√≥n
reporte = config_runtime.reporte.strip()
if not reporte.lower().endswith(".pbit"):
    reporte += ".pbit"

base_report_directory = config.base_report_directory

# ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî Buscar el archivo real en la carpeta ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
usuario = config_runtime.usuario
reporte = config_runtime.reporte.strip()
# Nos aseguramos de que termine en .pbit
if not reporte.lower().endswith(".pbit"):
    reporte += ".pbit"

# 1) Quitamos la extensi√≥n
raw = os.path.splitext(reporte)[0]   # e.g. "testUser + Empower BI V3 - Copy - Copy"

# 2) Si arranca con "usuario + ", lo eliminamos para quedarnos con el nombre real
prefijo = f"{usuario} + "
if raw.startswith(prefijo):
    nombre_real = raw[len(prefijo):]
else:
    nombre_real = raw

# 3) Construimos el √∫nico candidato: siempre "usuario + nombre_real.pbit"
candidato = f"{usuario} + {nombre_real}.pbit"

# 4) Buscamos en el directorio
files = os.listdir(base_report_directory)
if candidato in files:
    found_file = candidato
else:
    safe_print(f"‚ùå No encontr√© {candidato}; estos son los .pbit disponibles:")
    for f in files:
        if f.lower().endswith(".pbit"):
            safe_print(f"  ‚Äì {f}")
    sys.exit(1)

# 5) Ahora s√≠ tenemos el archivo correcto:
pbit_file_path = os.path.join(base_report_directory, found_file)
safe_print(f"üìä Analizando: {found_file}")

# ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî Construir y lanzar Power BI ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
powerbi_exe_path = config.powerbi_exe_path
command_line = f'"{powerbi_exe_path}" "{pbit_file_path}"'

safe_print("üü° Lanzando Power BI...")
app = Application(backend="uia").start(command_line)
safe_print("üü¢ Power BI lanzado.")
time.sleep(10)

# ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî Maximizar ventana de Power BI ‚Äî‚Äî‚Äî‚Äî‚Äî‚Äî
try:
    safe_print("üîç Buscando ventana principal de Power BI para maximizar...")
    max_window = None
    for _ in range(10):
        for w in Desktop(backend="uia").windows():
            if "Power BI" in w.window_text():
                max_window = w
                break
        if max_window:
            break
        time.sleep(1)

    if max_window:
        max_window.set_focus()
        max_window.maximize()
        safe_print("üü¢ Power BI maximizado correctamente.")
    else:
        safe_print("‚ö†Ô∏è No se encontr√≥ la ventana principal de Power BI para maximizar.")
except Exception as e:
    safe_print(f"‚ùå Error al maximizar Power BI: {e}")

time.sleep(1.5)
pyautogui.click(725, 1201)
```

```python
import pyperclip
import pyautogui
import time

for query in queries:
    print(f"\nüîÅ Ejecutando query: {query['name']}")

    # 1. Click en editor de DAX
    pyautogui.click(1551, 1320)  # O us√° el control [161] si est√°s con pywinauto
    time.sleep(0.3)
    pyautogui.hotkey('ctrl', 'a')
    pyautogui.press('backspace')

    # 2. Pegar el query
    pyperclip.copy(query['sql'])
    pyautogui.hotkey('ctrl', 'v')
    time.sleep(0.3)

    # 3. Ejecutar query
    pyautogui.click(1137, 1207)  # Bot√≥n "Run"
    print("[‚úì] Query ejecutado.")
    time.sleep(1.5)

    # 4. Manejar ventana 'Save As'
    print("[üìÅ] Esperando ventana 'Save As'...")
    save_as_window = None

    for child in window.descendants():
        if child.element_info.control_type == "Window" and "Save As" in child.window_text():
            save_as_window = child
            break

    if not save_as_window:
        print("‚ùå No se abri√≥ la ventana 'Save As'.")
        continue

    try:
        save_as_window.set_focus()
        print(f"[‚úì] Ventana 'Save As' detectada.")

        # 4.1 Escribir nombre del archivo (usando pyperclip para evitar errores con '+')
        nombreArchivo = f"{usuario}+{nombre_reporte}+{query['name']}.xlsx"
        pyperclip.copy(nombreArchivo)
        name_input = save_as_window.descendants()[113]
        name_input.set_focus()
        name_input.type_keys("^a{BACKSPACE}", with_spaces=True)
        pyautogui.hotkey("ctrl", "v")
        print(f"[‚úì] Nombre de archivo: {nombreArchivo}")

        # 4.2 Cambiar carpeta destino
        pyperclip.copy(carpeta_destino)
        pyautogui.hotkey("ctrl", "l")
        time.sleep(0.3)
        pyautogui.hotkey("ctrl", "v")
        pyautogui.press("enter")
        time.sleep(1.2)
        print(f"[‚úì] Carpeta destino: {carpeta_destino}")

        # 4.3 Revalidar ventana 'Save As'
        for child in window.descendants():
            if child.element_info.control_type == "Window" and "Save As" in child.window_text():
                save_as_window = child
                save_as_window.set_focus()
                break

        # 4.4 Click en bot√≥n Save
        save_button = save_as_window.descendants()[123]
        save_button.click_input()
        print("[‚úì] Click en 'Save' exitoso.")
        time.sleep(2)

        # 4.5 Confirmar reemplazo si aparece
        updated_children = window.descendants()
        yes_button = next(
            (c for c in updated_children if c.element_info.control_type == "Button" and c.window_text().strip().lower() == "yes"),
            None
        )
        if yes_button:
            yes_button.click_input()
            print("[‚Ü™] Confirmaci√≥n de reemplazo aceptada.")

        # 5. Verificar si se cerr√≥ correctamente
        time.sleep(1)
        final_check = any(
            "Save As" in c.window_text() and c.element_info.control_type == "Window"
            for c in window.descendants()
        )
        if not final_check:
            print(f"[‚úÖ] Archivo '{nombreArchivo}' guardado correctamente.")
        else:
            print(f"[‚ùå] Fall√≥ el guardado de '{nombreArchivo}'.")

    except Exception as e:
        print(f"‚ùå Error durante manejo de 'Save As': {e}")
```
